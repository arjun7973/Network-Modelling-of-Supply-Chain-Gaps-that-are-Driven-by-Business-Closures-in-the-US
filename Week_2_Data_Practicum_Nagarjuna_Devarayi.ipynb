# Data

**US Census.gov API**: Business Closures Data from the Business Dynamics Statistics (BDS) capturing data from 1978 - 2023 with 28 attributes on establishments and exits of firms across different states and territories (56), across sectors (19) and across different firm age categories (12). Total number of observations in the data = 534,888 rows.

```python
import pandas as pd
# Data API URL Link
url_bds = "https://www2.census.gov/programs-surveys/bds/tables/time-series/2023/bds2023_st_sec_fa.csv"
# Importing Data
df_bds = pd.read_csv(url_bds)
# Data Overview
df_bds.head()
```

```python
# Getting Data Information
df_bds.info()
```

**Data Types**: The data types for the numeric variables are set to object and require conversion to int type. Also, sector and fage variables require conversion to object type.

```python
# Feature Engineering: Sector Mapping from Codebook
sector_map = {
    '11': 'Agriculture, Forestry, Fishing and Hunting',
    '21': 'Mining, Quarrying, and Oil and Gas Extraction',
    '22': 'Utilities',
    '23': 'Construction',
    '31-33': 'Manufacturing',
    '42': 'Wholesale Trade',
    '44-45': 'Retail Trade',
    '48-49': 'Transportation and Warehousing',
    '51': 'Information',
    '52': 'Finance and Insurance',
    '53': 'Real Estate and Rental and Leasing',
    '54': 'Professional, Scientific, and Technical Services',
    '55': 'Management of Companies and Enterprises',
    '56': 'Administrative and Support and Waste Management and Remediation Services',
    '61': 'Educational Services',
    '62': 'Health Care and Social Assistance',
    '71': 'Arts, Entertainment, and Recreation',
    '72': 'Accommodation and Food Services',
    '81': 'Other Services (except Public Administration)'
}

# Convert sector to string first
df_bds['sector'] = df_bds['sector'].astype(str)

# Map section values
df_bds['sector_name'] = df_bds['sector'].map(sector_map)

# Feature Engineering: State Mapping from Codebook
state_map = {
    1: 'AL', 2: 'AK',
    4: 'AZ', 5: 'AR',
    6: 'CA', 8: 'CO',
    9: 'CT', 10: 'DE',
    11: 'DC', 12: 'FL',
    13: 'GA', 15: 'HI',
    16: 'ID', 17: 'IL',
    18: 'IN', 19: 'IA',
    20: 'KS', 21: 'KY',
    22: 'LA', 23: 'ME',
    24: 'MD', 25: 'MA',
    26: 'MI', 27: 'MN',
    28: 'MS', 29: 'MO',
    30: 'MT', 31: 'NE',
    32: 'NV', 33: 'NH',
    34: 'NJ', 35: 'NM',
    36: 'NY', 37: 'NC',
    38: 'ND', 39: 'OH',
    40: 'OK', 41: 'OR',
    42: 'PA', 44: 'RI',
    45: 'SC', 46: 'SD',
    47: 'TN', 48: 'TX',
    49: 'UT', 50: 'VT',
    51: 'VA', 53: 'WA',
    54: 'WV', 55: 'WI',
    56: 'WY', 72: 'PR'
}

# Mapping state values
df_bds['state_abbr'] = df_bds['st'].map(state_map)

# Fixing Data Type for Categorical Variables
categorical_cols = ['sector', 'fage', 'state_abbr', 'sector_name']

for col in categorical_cols:
    df_bds[col] = df_bds[col].astype('category')
```

```python
# Fixing Data Types for Numeric Variables
numeric_cols = [
    'firms',
    'estabs',
    'emp',
    'denom',
    'estabs_entry',
    'estabs_entry_rate',
    'estabs_exit',
    'estabs_exit_rate',
    'job_creation',
    'job_creation_births',
    'job_creation_continuers',
    'job_creation_rate_births',
    'job_creation_rate',
    'job_destruction',
    'job_destruction_deaths',
    'job_destruction_continuers',
    'job_destruction_rate_deaths',
    'job_destruction_rate',
    'net_job_creation',
    'net_job_creation_rate',
    'reallocation_rate',
    'firmdeath_firms',
    'firmdeath_estabs',
    'firmdeath_emp'
]

# Converting to numeric
df_bds[numeric_cols] = df_bds[numeric_cols].apply(
    pd.to_numeric,
    errors='coerce'
)
```

```python
# Rechecking the Data Types
df_bds.info()
```

**Missing Entries** Data has X and D entries which are identified from https://www.test.census.gov/content/dam/Census/programs-surveys/business-dynamics-statistics/codebook-glossary.pdf given as D = Disclosure Suppression and X = Structurally Missing

```python
# Setting X and D entries as NAs
import numpy as np
df_bds = df_bds.replace(['X', 'D'], np.nan)
df_bds.head()
```

```python
# Checking Missing Entries
df_bds.isnull().sum()
```

```python
# Checking percentge missing
(df_bds.isnull().sum() / len(df_bds)) * 100
```

The missing entries across each of the variables does not exceed 33%. Also, the missing entries are present only across the numeric variables and not the identifier and categorical variables of year, st (for state), sector, fage (for firm age), sector_name and sector_abbr

**Imputing**: Time Series Interpolation is applied below within the year, st, sector and fage combination of groups for the base variables with exception made for the derived variables which are computed from the base variables.

```python
# Sorting Data by st, sector, fage and year
df_bds = df_bds.sort_values(['st', 'sector', 'fage', 'year'])


# Getting Base Variables from Codebook
base_vars = [
    'firms',
    'estabs',
    'emp',
    'denom',
    'estabs_entry',
    'estabs_exit',
    'job_creation_births',
    'job_creation_continuers',
    'job_destruction_deaths',
    'job_destruction_continuers',
    'firmdeath_firms',
    'firmdeath_estabs',
    'firmdeath_emp'
]

# Time Series Interpolation of base variables
df_bds[base_vars] = (
    df_bds.groupby(['st', 'sector', 'fage'])[base_vars]
      .transform(lambda x: x.interpolate(limit_direction='both'))
)
```

```python
# Reconstructing the derived variables from base variables
# Job Flow Variables
df_bds['job_creation'] = (
    df_bds['job_creation_births'] +
    df_bds['job_creation_continuers']
)

df_bds['job_destruction'] = (
    df_bds['job_destruction_deaths'] +
    df_bds['job_destruction_continuers']
)

df_bds['net_job_creation'] = (
    df_bds['job_creation'] -
    df_bds['job_destruction']
)

# Rate Variables
df_bds['job_creation_rate_births'] = (
    100 * df_bds['job_creation_births'] / df_bds['denom']
)

df_bds['job_creation_rate'] = (
    100 * df_bds['job_creation'] / df_bds['denom']
)

df_bds['job_destruction_rate_deaths'] = (
    100 * df_bds['job_destruction_deaths'] / df_bds['denom']
)

df_bds['job_destruction_rate'] = (
    100 * df_bds['job_destruction'] / df_bds['denom']
)
df_bds['estabs_entry_rate'] = (
    100 * df_bds['estabs_entry'] / df_bds['estabs']
)

df_bds['estabs_exit_rate'] = (
    100 * df_bds['estabs_exit'] / df_bds['estabs']
)

df_bds['net_job_creation_rate'] = (
    df_bds['job_creation_rate'] -
    df_bds['job_destruction_rate']
)

df_bds['reallocation_rate'] = (
    df_bds['job_creation_rate'] +
    df_bds['job_destruction_rate'] -
    abs(df_bds['net_job_creation_rate'])
)
```

```python
# Rechecking for Missing Entries
df_bds.isnull().sum()
```

```python
# Checking percentge still missing
(df_bds.isnull().sum() / len(df_bds)) * 100
```

The imputation approach through time-series interpolation aimed at imputing only the base variables. However, when values are missing across successive time periods, the interplation returns NAs; this represents the case for base variables still containing missing entires. For the derived variables, where the base variables are not imputed or are 0, the derived variables is computed to NA.
The remaining missing entries across variables show that the missingness does not exceed 11%. Given the small percentage of missingness remaining, the records with missing entries were removed as below.

```python
# Removing 11% missing entries across columns
df_bds = df_bds.dropna()
```

**Federal Researve of the United States (FRED)**: Median Household Income by State from 1984-2024 containing 2050 entries.

```python
# Getting state Abbreviations
states = [
    'AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA',
    'HI','ID','IL','IN','IA','KS','KY','LA','ME','MD',
    'MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ',
    'NM','NY','NC','ND','OH','OK','OR','PA','RI','SC',
    'SD','TN','TX','UT','VT','VA','WA','WV','WI','WY'
]
# Initializing Empty Dataset for Storage
df_mhi = []
# Looping over states to get data for all states
for s in states:
    series_id = f"MEHOINUS{s}A672N"
    # FRED API url for Median Household Income
    url = f"https://fred.stlouisfed.org/graph/fredgraph.csv?id={series_id}"
    df = pd.read_csv(url)
    df['state'] = s
    df_mhi .append(df)
# Combining Datasets
df_mhi_full = pd.concat(df_mhi)
# Data Overview
df_mhi_full.head()
```

```python
# Checking Data info
df_mhi_full.info()
```

Data Cleaning: Since the datasets for each of the states was collected separately, the data required cleaning to bring all the median household income to one column.

```python
# Using date column as identifier column
id_col = 'observation_date'

# Melting all state columns into long format
df_long_mhi = df_mhi_full.melt(
    id_vars=[id_col],
    value_vars=[col for col in df_mhi_full.columns if col.startswith('MEHOINUS')],
    var_name='series_id',
    value_name='income'
)

# Removing missing rows
df_long_mhi = df_long_mhi.dropna(subset=['income'])
print(df_long_mhi.head())

# Extracting state abbreviation from series id
df_long_mhi['state_abbr'] = df_long_mhi['series_id'].str.extract(r'MEHOINUS([A-Z]{2})A672N')
df_long_mhi = df_long_mhi[
    ['observation_date', 'state_abbr', 'income']
].reset_index(drop=True)
print(df_long_mhi.head())
```

```python
# Feature Engineering: Creating year variable
df_long_mhi['observation_date'] = pd.to_datetime(df_long_mhi['observation_date'])
df_long_mhi['year'] = df_long_mhi['observation_date'].dt.year
# Dropping observation data
df_long_mhi = df_long_mhi.drop(columns=['observation_date'])
# Data Overview
df_long_mhi.head()
```

```python
# Checking Data Information
df_long_mhi.info()
```

**Federal Researve of the United States (FRED)**: Population by State from 1900-2025 containing 6200 entries.

```python
# Initializing empty dataframe for storage
df_pop = []
# Looping over states
for s in states:
    series_id = f"{s}POP"
    # FRED API url for Population Estimates
    url = f"https://fred.stlouisfed.org/graph/fredgraph.csv?id={series_id}"
    df = pd.read_csv(url)
    df.columns = ['observation_date', 'population']
    df['state'] = s
    df_pop.append(df)
# Combining Datasets
df_pop_full = pd.concat(df_pop, ignore_index=True)
# Data Overview
df_pop_full.head()
```

```python
# Checking Data info
df_pop_full.info()
```

```python
# Feature Engineering: Creating year variable
df_pop_full['observation_date'] = pd.to_datetime(df_pop_full['observation_date'])
df_pop_full['year'] = df_pop_full['observation_date'].dt.year
# Dropping observation data
df_pop_full = df_pop_full.drop(columns=['observation_date'])
# Data Overview
df_pop_full.head()
```

```python
# Checking Data Information
df_pop_full.info()
```

# Data Merging

Across all three data sets (Business Closures, Median Household Income and Population) the year and state variables are present. Inorder to preserve the firm age and sector variables, the Median Household Income and Population will be merged to the Business Closures dataset.

```python
# Getting common states across the three datasets to keep states and not territories
common_states = set(df_bds['state_abbr']) \
    .intersection(df_long_mhi['state_abbr']) \
    .intersection(df_pop_full['state'])

# Filtering the three datasets for common states
df_bds_filtered = df_bds[
    df_bds['state_abbr'].isin(common_states)
]

df_long_mhi_filtered = df_long_mhi[
    df_long_mhi['state_abbr'].isin(common_states)
]

df_pop_filtered = df_pop_full[
    df_pop_full['state'].isin(common_states)
]

# Merging the datasets by state and year
df_merged = df_bds_filtered.merge(
    df_long_mhi_filtered,
    on=['state_abbr', 'year'],
    how='left'
)

df_merged = df_merged.merge(
    df_pop_filtered.rename(columns={'state': 'state_abbr'}),
    on=['state_abbr', 'year'],
    how='left'
)

# Checking data info
df_merged.info()
```

```python
# Saving Datasets
df_bds.to_csv("df_bds.csv")
df_long_mhi.to_csv("df_long_mhi.csv")
df_pop_full.to_csv("df_pop_full.csv")
df_merged.to_csv("df_merged.csv")
```
